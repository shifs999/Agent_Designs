```python
import os
from groq import Groq
from dotenv import load_dotenv
from IPython.display import display_markdown
```

```python
load_dotenv()
client = Groq()
```

```python
generation_chat_history = [
    {
        "role": "system",
        "content": "You are a Python programmer tasked with generating high quality Python code."
        "Your task is to Generate the best content possible for the user's request. If the user provides critique," 
        "respond with a revised version of your previous attempt."
    }
]
```

```python
generation_chat_history.append(
    {
        "role": "user",
        "content": "Generate a Python implementation of the Depth First Search algorithm"
    }
)
```

```python
dfs_code = client.chat.completions.create(
    messages=generation_chat_history,
    model="llama3-70b-8192"
).choices[0].message.content

generation_chat_history.append(
    {
        "role": "assistant",
        "content": dfs_code
    }
)
```

```python
display_markdown(dfs_code, raw=True)
```

```python
reflection_chat_history = [
    {
    "role": "system",
    "content": "You are an experienced algorithm engineer, who develops and implements very structured and pragmatic algorithms. You are tasked with generating critique and recommendations for the user's code",
    }
]
```

```python
reflection_chat_history.append(
    {
        "role": "user",
        "content": dfs_code
    }
)
```

```python
critique = client.chat.completions.create(
    messages=reflection_chat_history,
    model="llama3-70b-8192"
).choices[0].message.content

display_markdown(critique, raw=True)
```

```python
generation_chat_history.append(
    {
        "role": "user",
        "content": critique
    }
)
```

```python
revised_res = client.chat.completions.create(
    messages=generation_chat_history,
    model="llama3-70b-8192"
).choices[0].message.content

display_markdown(revised_res, raw=True)
```

```python
generation_system_prompt = "You are a Python programmer tasked with generating structured, readable, and efficient Python code"

reflection_system_prompt = "You are an experienced algorithm engineer and computer scientist, who develops and implements very structured and pragmatic algorithms."
```

```python
def generate_response(user_msg, steps=10, verbose=1):
    from agentic_designs.reflection_des.reflection_agent_des import ReflectionAgent
    agent = ReflectionAgent()
    final_response = agent.run(
        user_msg=user_msg,
        generation_system_prompt=generation_system_prompt,
        reflection_system_prompt=reflection_system_prompt,
        n_steps=steps,
        verbose=verbose,
    )
    
    display_markdown(final_response, raw=True)
```

```python
generate_response("Generate a Python implementation of the Depth First Search algorithm")
```

```python
generate_response("Generate a Python implementation of the Mo's algorithm")
```
